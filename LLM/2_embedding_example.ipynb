```python
import os
import getpass

os.environ["OPENAI_API_KEY"] = getpass.getpass("OpenAI API Key:")
```

```python
from langchain.embeddings.openai import OpenAIEmbeddings

embedding = OpenAIEmbeddings()
```

```python
sentence1 = "the quick brown fox jumps over the lazy dog"
sentence2 = "i can see a fox jumping over a dog"
sentence3 = "the weather is wet and cold outside"
```

```python
embedding1 = embedding.embed_query(sentence1)
embedding2 = embedding.embed_query(sentence2)
embedding3 = embedding.embed_query(sentence3)
```

## Use an encoder to produce vector embeddings

Embeddings are numerical representations of the input text.

```python
print(f"{embedding1[:5]}\n{embedding2[:5]}\n{embedding3[:5]}")
```

```python
np.dot(embedding1, embedding2)
```

```python
np.dot(embedding1, embedding3)
```

```python
np.dot(embedding2, embedding3)
```

## Euclidean distance

Detailed explanation on [wikipedia](https://en.wikipedia.org/wiki/Euclidean_distance)

```python
def euclidean_distance(array1, array2):
    return np.sqrt(np.sum((array1 - array2) ** 2))
```

```python
array1 = np.array(embedding1)
array2 = np.array(embedding2)
array3 = np.array(embedding3)
```

```python
euclidean_distance(array1, array2)
```

```python
euclidean_distance(array1, array3)
```

```python
euclidean_distance(array2, array3)
```
